```python
aca_apiurl = "https://raw.githubusercontent.com/dmorton714/data-2-demo/main/states.csv"
aca = pd.read_csv(aca_apiurl)
aca.head()
```

```python
# 1.   Verify rows of data in the dataframes. 
print(len(surescripts.index))
print(len(aca.index))
print("---------------------")

# 2.   List the columns. 
print(surescripts.columns)
print(aca.columns)

for surescripts_col in surescripts.columns:
    print(surescripts_col)

for aca_col in aca.columns:
    print(aca_col)
print("---------------------")

# 3.   Find the types in the dataframes.
print(surescripts.info())
print(aca.info())
print("---------------------")

# 4.   Visualize the first 10 rows. 
print(surescripts.head(10))
print(aca.head(10))
print("---------------------")

# 5.   Find the data types. 
print(surescripts.dtypes)
print(aca.dtypes)
```

Convert column types to meaningful types 


1.   Convert one column to string
2.   Convert one column to float 
3.   Convert one column to bool 
4.   Convert one column to int
5.   Convert one column to datetime




# 1.   Convert one column to string

```python
surescripts['Region'] = surescripts['Region'].astype("string")
aca['State'] = aca['State'].astype("string")

print(surescripts.dtypes)
print(aca.dtypes)
```

# 2.   Convert one column to float 


```python
aca['Health Insurance Coverage Change (2010-2015)'] = aca['Health Insurance Coverage Change (2010-2015)'].astype(float)

print(aca.dtypes)
```

# 3.   Convert one column to bool 

```python
aca['State Medicaid Expansion (2016)'] = aca['State Medicaid Expansion (2016)'].astype('bool')

print(aca.dtypes)
```

# 4.   Convert one column to int

```python
aca['Marketplace Tax Credits (2016)'] = aca['Marketplace Tax Credits (2016)'].astype(int)

print(aca.dtypes)
```

# 5.   Convert one column to datetime

```python
surescripts['Period'] = pd.to_datetime(surescripts['Period'])

print(surescripts.dtypes)
```
